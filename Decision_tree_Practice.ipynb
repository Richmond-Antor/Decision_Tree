```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Cat-and-Dog/shop data.csv')
```

```python
df.shape
```

```python
df
```

```python
y = df.buys
```

```python
y
```

```python
x = df.drop('buys', axis=1)
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_x= LabelEncoder()
x = x.apply(le_x.fit_transform)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.tree import DecisionTreeClassifier
dect = DecisionTreeClassifier()
dect.fit(x_train, y_train)
```

```python
dect.predict(x_test)
```

```python
y_test
```

```python
dect.score(x_test, y_test)
```

```python
#Cross validation score

from sklearn.model_selection import cross_val_score
```

```python
result = cross_val_score(DecisionTreeClassifier(), x, y, cv=5)
result
```

```python
result.mean()
```

```python
from sklearn.linear_model import LogisticRegression
result = cross_val_score(LogisticRegression(), x, y, cv=5)
result
```

```python
result.mean()
```

```python
from sklearn.preprocessing import MinMaxScaler
MMS = MinMaxScaler()
x_scaled = MMS.fit_transform(x)
x_scaled
```

```python
from sklearn.neighbors import KNeighborsClassifier
result = cross_val_score(KNeighborsClassifier(n_neighbors=3), x_scaled, y, cv=5)
result
```

```python
result.mean()
```
